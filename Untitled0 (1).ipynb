```python
!pip install PyMuPDF
import fitz

# Function to extract text from specific pages of a PDF
def extract_pages(pdf_path, start_page, num_pages):
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(start_page, start_page + num_pages):
        text += doc[page_num].get_text()
    doc.close()
    return text

pdf_path = '/content/Harry_Potter_(www.ztcprep.com).pdf'  # Replace with the actual path to your PDF file
```

```python
# prompt: Write Python code and use MapReduct to count occurrences of each word in the first text file (file.txt). How many times each word is repeated

from collections import defaultdict

def mapper(filename):
  """Maps words in a file to their counts."""
  word_counts = defaultdict(int)
  with open(filename, 'r', encoding='utf-8') as f:
    for line in f:
      for word in line.strip().split():
        word_counts[word] += 1
  for word, count in word_counts.items():
    yield (word, count)

def reducer(word, counts):
  """Reduces the word counts from multiple mappers."""
  total_count = sum(counts)
  yield (word, total_count)


if __name__ == '__main__':
  filename = 'file1.txt'

  # Map phase:
  mapped_results = []
  for result in mapper(filename):
    mapped_results.append(result)

  # Group by key (word)
  grouped_results = defaultdict(list)
  for word, count in mapped_results:
    grouped_results[word].append(count)

  # Reduce phase:
  reduced_results = []
  for word, counts in grouped_results.items():
    for result in reducer(word, counts):
      reduced_results.append(result)


  # Print the results
  for word, count in reduced_results:
    print(f'{word}: {count}')
```

```python
# prompt:  From the second text file (file2.txt), write Python code and use MapReduct to count how many times non-English words (names, places, spells etc.) were used. List those words and how many times each was repeated.

from collections import defaultdict

def mapper(filename):
  """Maps words in a file to their counts."""
  word_counts = defaultdict(int)
  with open(filename, 'r', encoding='utf-8') as f:
    for line in f:
      for word in line.strip().split():
        # Assuming non-English words contain characters not in the English alphabet
        if not word.isalpha() or not word.isascii():
          word_counts[word] += 1
  for word, count in word_counts.items():
    yield (word, count)

def reducer(word, counts):
  """Reduces the word counts from multiple mappers."""
  total_count = sum(counts)
  yield (word, total_count)


if __name__ == '__main__':
  filename = 'file2.txt'

  # Map phase:
  mapped_results = []
  for result in mapper(filename):
    mapped_results.append(result)

  # Group by key (word)
  grouped_results = defaultdict(list)
  for word, count in mapped_results:
    grouped_results[word].append(count)

  # Reduce phase:
  reduced_results = []
  for word, counts in grouped_results.items():
    for result in reducer(word, counts):
      reduced_results.append(result)

  # Print the results
  for word, count in reduced_results:
    print(f'{word}: {count}')
```
